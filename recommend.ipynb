```python
columns_to_drop = ["Launch Date", "Participants (Course Content Accessed)", "Audited (> 50% Course Content Accessed)", "% Male", "% Female", "Median Hours for Certification", "Course Number", "Year", "Honor Code Certificates", "Total Course Hours (Thousands)"]


columns_to_drop += data.columns[10:17].tolist()



courses = data.drop(columns=columns_to_drop)

courses.drop(["Median Age", "% Bachelor's Degree or Higher"], axis=1, inplace=True)
```

```python
courses.columns
```

```python
courses = courses.dropna()
```

```python
tokenz = lambda x:x.split()
courses["Instructors"] = courses["Instructors"].apply(lambda x:x.split(","))
courses["Course Subject"] = courses["Course Subject"].apply(tokenz)
```

```python
courses
```

```python
courses["Instructors"]=courses["Instructors"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
courses["tags"] = courses["Course Subject"] + courses["Instructors"]
print(len(courses.iloc[0]["tags"]))
```

```python
def append_string(row):
    return row['tags'] + [row['Institution']]

courses["tags"] = courses.apply(append_string, axis=1)
```

```python
new_df = courses[["Course Title", "Instructors", "tags"]]
```

```python
new_df["tags"] = (new_df["tags"].apply(lambda x: " ".join(x)).replace(',', "")).apply(lambda x: x.replace(",", ""))
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer 

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df['tags'].apply(stem)
```

```python
path = "C:\\Users\\Lenovo\\Desktop\\practice+projex\\CodeCraft-TM-PAI\\CodeCraft-TM-PAI\\new_df.csv"
new_df.to_csv(path)
```

```python
vectors =  tfidf.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = new_df[new_df['Course Title'] == course].index[0]
    course_list = sorted(list(enumerate(similarity[course_index])), key=lambda x: x[1], reverse=True)
    list_ = []
    for i in course_list:
      if new_df.iloc[i[0]]["Course Title"] == course:
         continue
      if len(list(set(list_)))<6:
         list_.append(new_df.iloc[i[0]]["Course Title"])
    #print(f"{type(new_df.iloc[i[0]]['Course Title'])}")
       
    return list(set(list_))
       

```

```python
a = recommend("Circuits and Electronics")
display(a)
```

```python
import hashlib

def sha256_hash(text):
    # Encode the text into bytes before hashing
    text_bytes = text.encode('utf-8')
    # Create a new SHA-256 hash object
    sha256_hash_obj = hashlib.sha256()
    # Update the hash object with the bytes
    sha256_hash_obj.update(text_bytes)
    # Get the hexadecimal digest of the hash
    hashed_text = sha256_hash_obj.hexdigest()
    return hashed_text

# Strings to hash
strings = [
    "abefgh",
    "abc",
    "acdefg1",
    "abcde1fg1",
    "cdefg1",
    "abc1efg1",
    "ab1cd1efg1",
    "cefg",
    "abcg1",
    "abcfg1",
    "abcdefg10000",
    "abefg11",
    "abcdefg1119999",
    "abcd987efg1",
    "abcdefg111",
    "abcdefg",
    "abc67defg1111",
    "abcdefg1",
    "abcde898989fg",
    "abcdefg1111"
]

# Hash each string using SHA-256
for text in strings:
    hashed_text = sha256_hash(text)
    print(f"Text: {text}\nHash: {hashed_text}\n")
```
